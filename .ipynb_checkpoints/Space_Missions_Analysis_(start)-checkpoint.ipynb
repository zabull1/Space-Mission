# Use a Choropleth Map to Show the Number of Launches by Country

* Create a choropleth map using [the plotly documentation](https://plotly.com/python/choropleth-maps/)
* Experiment with [plotly's available colours](https://plotly.com/python/builtin-colorscales/). I quite like the sequential colour `matter` on this map. 
* You'll need to extract a `country` feature as well as change the country names that no longer exist.

Wrangle the Country Names

You'll need to use a 3 letter country code for each country. You might have to change some country names.

* Russia is the Russian Federation
* New Mexico should be USA
* Yellow Sea refers to China
* Shahrud Missile Test Site should be Iran
* Pacific Missile Range Facility should be USA
* Barents Sea should be Russian Federation
* Gran Canaria should be USA


You can use the iso3166 package to convert the country names to Alpha3 format.

```python
# df_data.drop('Country', axis = 1, inplace =True)
# df_data
```

```python
df_data['Country'] = df_data.Location.astype('str').str.split(',').str.get(-1)
display(df_data)
countries = df_data.groupby('Country').count()
display(countries)
```

```python
for country in countries.index:
    if country.strip() == 'Kazakhstan':
        df_data.Country = df_data.Country.astype('str').str.replace(country,' Russia')
    elif country.strip() == 'Barents Sea':
        df_data.Country = df_data.Country.str.replace(country,' Russia')
    elif country.strip() == 'New Mexico':
         df_data.Country = df_data.Country.str.replace(country, ' USA')
    elif country.strip() == 'Pacific Missile Range Facility':
         df_data.Country = df_data.Country.str.replace(country, ' USA')
    elif country.strip() == 'Gran Canaria':
         df_data.Country = df_data.Country.str.replace(country, ' USA')
    elif country.strip() == 'Pacific Ocean':
         df_data.Country = df_data.Country.str.replace(country, ' USA')        
    elif country.strip() == 'Yellow Sea':
         df_data.Country = df_data.Country.str.replace(country, ' China')
    elif country.strip() == 'Shahrud Missile Test Site':
         df_data.Country = df_data.Country.str.replace(country, ' Iran')

df_data.groupby('Country').count()
```

```python
from iso3166 import countries as cs

# for c in cs:
#     for i in range(4324):
#         if df_data.iloc[i].Country == c.name:
#             print(c.name)
#     if country.strip() == 'Kazakhstan':
# df_data.Country = df_data.Country.str.replace(country, ' Iran')

!pip install pycountry

import pycountry

def do_fuzzy_search(country):
    try:
        result = pycountry.countries.search_fuzzy(country)
        return result[0].alpha_3
    except:
        return np.nan
    
df_data["ISO"] = df_data["Country"].apply(lambda country: do_fuzzy_search(country))

df_data
```

```python
df_ISO = df_data.groupby(['ISO', 'Country'], as_index=False).count()
df_ISO
```

```python
fig = px.choropleth(df_ISO,
                    locations = 'ISO',
                    color='Organisation',
                    hover_name = 'Country',
                    color_continuous_scale = px.colors.sequential.matter)

fig.update_layout(
        coloraxis_showscale=False,
        title= 'Map of World')
fig
```

## Chart the Percentage of Failures over Time

Did failures go up or down over time? Did the countries get better at minimising risk and improving their chances of success over time? 

```python
df_failed_missions = df_data.query("Mission_Status != 'Success'").groupby('year',as_index= False).count()[['year','Organisation']].rename(columns = {"Organisation":"Failed_Missions"})
df_succesful_missions = df_data.query("Mission_Status == 'Success'").groupby('year',as_index= False).count()[['year','Organisation']].rename(columns = {"Organisation":"Successful_Missions"})
#df_data.groupby(['year','Mission_Status'], as_index = False).count()
df_successful_failed_missions = pd.merge(df_succesful_missions, df_failed_missions)
df_successful_failed_missions
```

```python
df_successful_failed_missions['Total_Missions'] = df_successful_failed_missions['Successful_Missions'] + df_successful_failed_missions['Failed_Missions']
df_successful_failed_missions
```

```python
df_successful_failed_missions['percent_Successful_Missions'] = (df_successful_failed_missions['Successful_Missions']/df_successful_failed_missions['Total_Missions'])*100
df_successful_failed_missions
```

```python
df_successful_failed_missions['percent_Failed_Missions'] = 100- df_successful_failed_missions['percent_Successful_Missions']
df_successful_failed_missions
```

```python
bar_percent_failed = px.bar(x=df_successful_failed_missions.year, 
                            y= df_successful_failed_missions.percent_Failed_Missions,
                           color = df_successful_failed_missions.percent_Failed_Missions )
bar_percent_failed.update_layout(xaxis_title = 'Year',
                                yaxis_title = 'Percentage of Failures',
                                title = 'Percentage of Failures Over time',
                                coloraxis_showscale = False)
```
